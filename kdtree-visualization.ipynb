# Visualize k-d tree

This notebook shows visualisation of k-d tree 

In the first cell, we generate `n` points and run kd tree algoritms. The second cell shows a non-interactive step-by-step visualisation. The last cell shows the same visualisation but you can jump randomly in interations.

```python
%matplotlib notebook  
import kdtree
import numpy as np
import drawing

def generate_points(n, max_value):
    return np.random.randint(max_value,size=(n,2))

max_value = 1000
n = 60
pts = generate_points(n,max_value)

tree = kdtree.nnkdtree(pts)
points = kdtree.collect_points(tree)
lines = kdtree.collect_lines(tree)
```

```python
drawing.draw_static_animation(points,lines)    
```

```python
drawing.draw_interactive_animation(points,lines)  
```
